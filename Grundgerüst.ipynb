## 3. Statistik der Rohdaten (Analyse im Long-Format)
1. *Art der Variable:*
Die Gesamtfertilitätsrate (TFR) ist ein Durchschnittswert. Sie wird daher statistisch als kontinuierliche Variable behandelt, nicht als diskret.
2. *Anzahl Datenpunkte:*
Im Long-Format sind Hunderte bis Tausende von Datenpunkten (Länder über Jahre) verfügbar. Es gibt jedoch Datenlücken (Missing Values), insbesondere in älteren Reihen und in Entwicklungsländern.
3. *Verteilung der zu untersuchenden Variable(n):*
Die globale Verteilung ist oft linksschief oder bimodal(Histogramme). Dies liegt an vielen niedrigen Raten in Industrieländern und wenigen sehr hohen Raten in afrikanischen Ländern.

4. *Zusammenhänge zwischen zwei Variablen:*
Es besteht ein starker negativer Zusammenhang mit anderen Variablen. Steigender Wohlstand (BIP) und höheres Bildungsniveau von Frauen gehen mit einer sinkenden TFR einher.

## 4. Analyse 





#### "Welche Länder verzeichneten die stärksten jährlichen Rückgänge der Geburtenraten?"
 

```python
import pandas as pd
import matplotlib.pyplot as plt
df_pivot_asc = df_pivot.sort_index(ascending=True)
start_rate = df_pivot_asc.iloc[0] 
end_rate = df_pivot_asc.iloc[-1] 
total_change_all_periods = end_rate - start_rate
strongest_declines = (total_change_all_periods * -1).sort_values(ascending=False).head(10)
result_df_new = pd.DataFrame(strongest_declines).rename(columns={0: 'Gesamter Rückgang (TFR-Punkte)'})
print("Top 10 stärkste Rückgänge (über den längsten verfügbaren Zeitraum):")
result_df_new
```

```python
plt.figure(figsize=(12, 6))
strongest_declines.plot(kind='bar')
plt.title("Top 10 stärkste Rückgänge der Geburtenrate")
plt.xlabel("Land/Gebiet")
plt.ylabel("Gesamter Rückgang der TFR (Punkte)")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
